```python
# Import dependencies
import random
from citipy import citipy
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
# Create empty lists to store latitudes and city_names for later use
city_names = []

# Generate random floats between full range of latitudes and longitudes using random.uniform()
# Use while loop to get 500 *unique* city names using set() function at the end of the loop to delete duplicates

while len(city_names) < 650:
    latitude = random.uniform(-90, 90)
    longitude = random.uniform(-180, 180)
    
    city = citipy.nearest_city(latitude, longitude)
    city_names.append(city.city_name)
        
    city_names = list(set(city_names))
```

```python
# Create empty strings to store data for temperature, humidity, cloudiness, wind speed, and latitude
temperatures = []
humidities = []
cloudiness_list = []
wind_speeds = []
latitudes = []
cities = []
```

```python
# API Calls from OpenWeatherMap API
api_key = "&appid=9a52d7caa9fc25c09c68c09ee9a0701a"
print("Beginning Data Retrieval")
print("------------------------------------")

for i in range(650):
    city_name = city_names[i]
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city_name}{api_key}&units=imperial"
    print(f"Processing Record {i+1} of 650 | {city_name}")
    print(url)
    weather_data = requests.get(url).json()
    try:
        city_name = weather_data["name"]
        latitude = weather_data["coord"]["lat"]
        temperature = weather_data["main"]["temp_max"]
        humidity = weather_data["main"]["humidity"]
        cloudiness = weather_data["clouds"]["all"]
        wind_speed = weather_data["wind"]["speed"]
           
        temperatures.append(temperature)
        humidities.append(humidity)
        cloudiness_list.append(cloudiness)
        wind_speeds.append(wind_speed)
        latitudes.append(latitude)
        cities.append(city_name)
    except KeyError:
        print("Oops! No city with that name was found...")
```

```python
# Create weather data frame from lists
weather_df = pd.DataFrame({"City": cities,
                          "Max Temperature (F)": temperatures,
                          "Humidity (%)": humidities,
                          "Cloudiness (%)": cloudiness_list,
                          "Wind Speed (mph)": wind_speeds,
                          "Latitude": latitudes})
weather_df.to_csv("Created Files/CityWeatherData.csv")
weather_df.head()
```

```python
# Create scatter plot for latitude vs. temperature
plt.scatter(latitudes,temperatures, c='b', linewidths=1, edgecolors='black')
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title("City Latitude vs. Max Temperature")
plt.grid(color="grey", alpha=.25)
plt.savefig("Created Files/LatVsTemp.png")
plt.show()
```
